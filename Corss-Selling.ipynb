<h2> Business Objective: </h2>

Identify the opportunities for cross- selling in a retail store    

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from tidypython import *
```

```python
data=pd.read_excel('transactions_by_dept.xlsx')
data.head()
```

```python
data['Dept'].nunique()
```

```python
data['POS Txn'].nunique()
```

The items sold in the store are:

```python
data['Dept'].unique()
```

```python
relevantdata=data[['POS Txn', 'Dept','Sales U']]
relevantdata.head()
```

```python

clean_data = relevantdata.groupby(['POS Txn', 'Dept'], as_index=False).agg({
    'Sales U': 'sum'
})
clean_data = (clean_data.groupby(['POS Txn', 'Dept'])['Sales U']
          .sum().unstack().reset_index().fillna(0)
          .set_index('POS Txn'))
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

clean_data = clean_data.applymap(encode_units)
```

```python
clean_data.head()
```

<h6>  Generating frequent itemsets and association rules for a recommender system using
Association Rule Mining :
    

Confidence: This says how likely item Y is purchased when item X is purchased, expressed as {X -> Y}. This is measured by the proportion of transactions with item X, in which item Y also appears. In Table 1, the confidence of {apple -> beer} is 3 out of 4, or 75%.

Support: This says how popular an itemset is, as measured by the proportion of transactions in which an itemset appears.


Used support as 1% & confidence as 20% as higher values yield no/very less results

```python
frequent_itemsets = apriori(clean_data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules


association_rules(frequent_itemsets, metric='confidence', min_threshold=0.2)
```

<h6> Recommendations </h6>


1.Wine should be placed near american greetings & candy: As there is lower confidence associated with these pair of products, we need to promote these products as a basket by placing them next to each other.Customers who have the intention of gifting will then consider complementary products like greetings, candy or wine.
For example: I decide to purchase wine as a gift for a friend. Since, wine is next to greetings section, I might also consider purchasing a greeting card and a fancy gift cover for the wine bottle.

2.Personal Care & Household cleaning products should be placed furthest from & Health Care products : Since there is a high confidence, shoppers who tend to purchase one category of products have high probability of purchasing the other category.
Therefore, by placing them at a distance we can ensure that the shopper has to travel Personal Care & Household cleaning  aisle to  Health Care aisle and is subsequently exposed to other products in the store.

3.Beer, Spirits & Tobacco should be placed next to each other: As there is lower confidence associated with these pair of products, we need to promote these products as a basket by placing them next to each other. 86% of smokers drink alcohol (according to National Center for Biotechnology Information*), therefore placing them together will increase cutomers' convenience and boost sales.
